```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
b = 20
a = 0.1

N = 10_000_000
x = np.random.random(N)
s = x > 0.5
x = (2*x-s)**(1/(2*b))

x *= (2*s-1)
x += 2*(1-s)-1
x *= np.pi
x -= a

plt.figure()
plt.hist(x, bins=500, density=True)

t = np.linspace(-np.pi, +np.pi, 1_000)
plt.plot(t, np.abs(np.sinc(b*(a+np.sin(t))))*b/np.pi)
plt.show()
```

```python
N = 10_000_000
x = np.random.random(N)
n = 5

s = x > 0.5
x = 2*x-s

x = 1 - x**(1/(n+1))
x = np.pi * x*(2*s-1)

plt.figure()
plt.hist(x, bins=500, density=True)

t = np.linspace(-np.pi, +np.pi, 1_000)
plt.plot(t, (1-np.abs(t)/np.pi)**n * (n+1) / (2*np.pi))
plt.ylim(0, 3)
plt.show()
```
