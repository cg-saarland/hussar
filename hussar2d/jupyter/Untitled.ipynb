```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tx_x = 0
tx_y = 1

rx_x = 0
rx_y = 1

k0 = 200

x = np.linspace(0, 1, 1_000)
tx_d = np.sqrt(tx_y**2 + (tx_x-x)**2)
rx_d = np.sqrt(rx_y**2 + (rx_x-x)**2)

v = np.exp(1j*k0 * (tx_d + rx_d)) / (np.sqrt(tx_d) * np.sqrt(rx_d))

def normWithYEquals1(x):
    return x / np.sqrt(x**2 + 1)

wi_x = normWithYEquals1(rx_x - x)
wi_y = normWithYEquals1(tx_x - x)
dddu = wi_x + wi_y

K = 1e-2
vs = np.sinc(-k0/np.pi*dddu*K)

plt.figure(figsize=(16,5))
plt.plot(x, np.abs(v))
plt.plot(x, np.real(v))
plt.show()

vfiltered = v * vs

print(np.mean(v), np.mean(vfiltered))

plt.figure(figsize=(16,5))
plt.plot(x, np.abs(vfiltered))
plt.plot(x, np.real(vfiltered))
plt.show()

plt.figure(figsize=(16,5))
plt.plot(x, np.abs(vs))
plt.plot(x, np.real(vs))
plt.show()
```
