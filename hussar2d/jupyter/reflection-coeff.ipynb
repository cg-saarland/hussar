```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
S = 50
x = np.linspace(-S, +S, 1_000_000)
r = np.sqrt(1 + x**2)

k0 = 40

v = 1j*np.sqrt(k0) * np.exp(1j*k0 * r) / np.sqrt(r)
v = v**2

v /= np.sqrt(k0) * np.sqrt(2)

plt.figure()
plt.plot(x, np.abs(v))
plt.show()

print(np.abs(np.mean(v) * 2*S))
```
